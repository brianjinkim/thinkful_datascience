```python
#for log transforming the non normally distributed data
def getLog(x):
    if x == 0:
        return 0
    return np.log(x)
```

```python
#bool function for high 0 columns
def getBool(x):
    if x == 0:
        return 0
    return 1
```

```python
#convert df to log features
transform = pd.DataFrame()
for col in df.iloc[:, 1:]:
    transform[col] = df[col].apply(getLog)

#features with high 0 count convert to bool
for col in df[['murder', 'rape', 'robbery', 'motor_theft', 'arson']]:
    transform[col] = df[col].apply(getBool)

sns.pairplot(transform)
```

## standardize and reduce/pca

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import warnings
warnings.filterwarnings('ignore')
```

```python
df_pca = features

# Normalize the data so that all variables have a mean of 0 and standard deviation of 1.
X = StandardScaler().fit_transform(df_pca)

# The NumPy covariance function assumes that variables are represented by rows not columns, so we transpose X.
Xt = X.T
Cx = np.cov(Xt)
```

```python
# Calculating eigenvalues and eigenvectors.
sns.set_style('whitegrid')
eig_val_cov, eig_vec_cov = np.linalg.eig(Cx)

plt.plot(eig_val_cov)
plt.show()
```

```python
#check eigenvalues and eigenvectors
for i in range(len(eig_val_cov)):
    eigvec_cov = eig_vec_cov[:, i].reshape(1, -1).T
    print('Eigenvector {}: \n{}'.format(i+1, eigvec_cov))
    print('Eigenvalue {}: {}'.format(i+1, eig_val_cov[i]))
    print(40*'-')
    
print('Percentage of total variance in dataset explained by each component calculated by hand.\n', eig_val_cov / sum(eig_val_cov)) 
```

```python
# use first eigen vector to create P. It will be used to tranform Cx
# into Cy to get Y, the dimensionally-reduced representation of X.
P = eigvec_cov[:, 0]

# Transform X into Y.
Y = P.T.dot(Xt)

# Combine X and Y for plotting purposes.
df_pca['Component'] = Y
data_plot = pd.melt(df_pca, id_vars='Component')

g = sns.FacetGrid(data_plot, col="variable", size=6, aspect=.5)
g.map(
    sns.regplot,
    "Component",
    "value",
    x_jitter=.5,
    y_jitter=.5,
    fit_reg=False
)
plt.show()
```

```python
import statsmodels.api as sm
```

```python
# get relationship of predictors - motor_theft
# Bivariate relationships of predictors to admissions outcome.
sns.boxplot(x="motor_theft", y="population", data=features)
plt.show()

# Relationship between continuous data
#sns.regplot(x='', y='population', data=df)
#plt.show()
```

```python
# Declare predictors.
X_statsmod = features[['population']]

# The Statsmodels formulation requires a column with constant value 1 that will act as the intercept.
X_statsmod['intercept'] = 1 

# Declare and fit the model.
logit = sm.Logit(features['motor_theft'], X_statsmod)
result = logit.fit()

# Lots of information about the model and its coefficients, but the
# accuracy rate for predictions is missing.
print(result.summary())
```

```python
# Calculate accuracy. First, get probability that each row will be admitted.
pred_statsmod = result.predict(X_statsmod)

# Code admission as 1 if probability is greater than .5.
pred_y_statsmod = np.where(pred_statsmod < .5, 0, 1)

# Accuracy table.
table = pd.crosstab(features['motor_theft'], pred_y_statsmod)

print('\n Accuracy by motor theft')
print(table)
print('\n Percentage accuracy')
print((table.iloc[0,0] + table.iloc[1,1]) / (table.sum().sum()))
```

```python
features.drop('Component', 1, inplace=True)
features.head()
```

```python
from sklearn.model_selection  import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
y = features['motor_theft']
X = features.loc[:, ~(features.columns).isin(['motor_theft'])]

# Create Training Set and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
# Declare a logistic regression classifier.
# Parameter regularization coefficient C described above.
lr = linear_model.LogisticRegression(penalty='l1', C=10)

# Fit the model.
fit = lr.fit(X, y)

# Display.
print('Coefficients')
print(fit.coef_)
print(fit.intercept_)
pred_y_sklearn = lr.predict(X)

print('\n Accuracy by motor theft')
print(pd.crosstab(pred_y_sklearn, y))

print('\n Percentage accuracy')
print(lr.score(X, y))
```

```python
# Create new training Set and testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4)
```

```python
print('\n Percentage accuracy')
print('score:', lr.score(X_test, y_test))
print('10-fold CV:', cross_val_score(lr, X_test, y_test, cv=10))
```

```python
ridgeregr = linear_model.Ridge(alpha=10, fit_intercept=True) 

fit = ridgeregr.fit(X_train, y_train)

# Display.
print('Coefficients')
print(fit.coef_)
print(fit.intercept_)

print('\n Percentage accuracy')
print('score:', ridgeregr.score(X_test, y_test))
print('10-fold CV:', cross_val_score(ridgeregr, X_test, y_test, cv=10))
```

```python
lass = linear_model.Lasso(alpha=0.05) 
fit = lass.fit(X_train, y_train)

# Display.
print('Coefficients')
print(fit.coef_)
print(fit.intercept_)


print('R² for the model with few features:')
print('score:', lass.score(X_test, y_test))
print('10-fold CV:', cross_val_score(lass, X_test, y_test, cv=10))
print("-" * 100 + '\n')
```
