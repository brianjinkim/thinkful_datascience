```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#http://archive.ics.uci.edu/ml/datasets/Adult
income_raw = pd.read_csv('../Datasets/finance.csv', header=None)
income_raw.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'martial_status', 'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'salary']
df = income_raw
df.head()
```

```python
df.info()
```

```python
#strip white spaces in data
def white_del(x):
    return df[x].str.strip()

df['workclass'] = white_del('workclass')
df['martial_status'] = white_del('martial_status')
df['occupation'] = white_del('occupation')
df['relationship'] = white_del('relationship')
df['race'] = white_del('race')
df['sex'] = white_del('sex')
df['native_country'] = white_del('native_country')
df['salary'] = white_del('salary')
```

```python
#change columns into categories
def race_id(x):
    if x == 'White':
        return 1
    elif x == 'Asian-Pac-Islander':
        return 2
    elif x == 'Amer-Indian-Eskimo':
        return 3
    elif x == 'Other':
        return 4
    elif x == 'Black':
        return 5
    return 0
    
def relationship_id(x):
    if x=='Wife' or x=='Husband' or x=='Own-child':
        return 1
    else:
        return 0

def workclass_id(x):
    if x == 'Private':
        return 1
    elif x == 'Self-emp-not-inc' or x=='Self-emp-inc':
        return 2
    elif x=='Federal-gov' or x=='Local-gov' or x=='State-gov':
        return 3
    return 0

def married_id(x):
    if x == 'Married-civ-spouse' or x == 'Married-spouse-absent' or x == 'Married-AF-spouse':
        return 1
    return 0

def occupation(x):
    if x == 1:
        return 0
    return 1
    

df['male'] = np.where(df['sex'] == 'Male', 1, 0)
df['is_above_50k'] = np.where(df['salary'] == '<=50K', 0, 1)
df['race_id'] = df.race.apply(race_id)
df['has_family'] = df.relationship.apply(relationship_id)
df['workclass_id'] = df.workclass.apply(workclass_id)
df['is_married'] = df.martial_status.apply(married_id)
```

```python
df.head()
```

```python
#drop columns
df.drop(['relationship', 'race', 'sex', 'salary', 'fnlwgt', 'workclass', 'education', 'martial_status', 'occupation'], axis=1,  inplace=True)
df.head()
```

```python
df.corr()
```

```python
# This is the model we'll be using.
from sklearn import tree

# A convenience for displaying visualizations.
from IPython.display import Image

# Packages for rendering our tree.
import pydotplus
import graphviz

# Package to take time
import time

# Initialize and train our tree.
decision_tree = tree.DecisionTreeClassifier(
    criterion='entropy',
    max_features=1,
    max_depth=4
)

start_time = time.time()
decision_tree.fit(df.drop(['is_above_50k', 'native_country'], 1), df['is_above_50k'])
print("--- %s seconds ---" % (time.time() - start_time))

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=df.drop(['is_above_50k', 'native_country'], 1).columns,
    class_names=['Makes Below 50K', 'Makes Above 50K'],
    filled=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
from sklearn import ensemble
from sklearn.model_selection import cross_val_score

rfc = ensemble.RandomForestClassifier()

start_time = time.time()
rfc.fit(df.drop(['is_above_50k', 'native_country'], 1), df['is_above_50k'])
print("--- %s seconds ---" % (time.time() - start_time))

# Render our tree.
dot_data = tree.export_graphviz(
    decision_tree, out_file=None,
    feature_names=df.drop(['is_above_50k', 'native_country'], 1).columns,
    class_names=['Makes Below 50K', 'Makes Above 50K'],
    filled=True
)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```
