```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#styling
sns.set_style('white')

df = pd.read_csv('../Datasets/ESSdata_Thinkful.csv')
print(df.head())
print(df.cntry.unique())
```

```python
# Making dummies.
# Also creating a data frame to hold our features.
features = pd.get_dummies(df['cntry'])

# Nordic countries dummy.
features['Nordic'] = np.where((df['cntry'].isin(['NO', 'SE'])), 1, 0)

print(pd.crosstab(features['Nordic'], df['cntry']))
```

```python
#take continuous variable and make it categorical

features['Adult_65plus'] = np.where(df['agea']>=65, 1, 0)

# Check to make sure it worked.
print(df['agea'].groupby(features['Adult_65plus']).describe())
```

```python
corrmat = df.corr()

#setup the matplotlib figure
f, ax = plt.subplots(figsize=(12,9))

#draw heatmap using seaborn
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()

#pplfair, ppltrst, and pplhlp are redundant with each other. Lets make
#averages. In case some variables are on different scales than others
#(IE 1 ranges from 0 - 100, 1 ranges from 0-7) we scale it before
#averaging by subtracting the average of each variable from all values
#in that variable, then dividing by the standard deviation

means = df[['ppltrst', 'pplfair', 'pplhlp']].mean(axis = 0)
stds = df[['ppltrst', 'pplfair', 'pplhlp']].std(axis=0)
features['trust_fair_help'] = ((df[['ppltrst', 'pplfair', 'pplhlp']] - means) / stds).mean(axis=1)

#check how the composite correlates with each of the individual variables
plotdf = df.loc[:, ['ppltrst', 'pplfair', 'pplhlp']]
plotdf['trust_fair_help'] = features['trust_fair_help']
corrmat2 = plotdf.corr()

print(corrmat2)
```

```python
#make four panel plot
fig = plt.figure()

fig.add_subplot(221)
plt.hist(df['agea'].dropna())
plt.title('Raw')

fig.add_subplot(222)
plt.hist(np.log(df['agea'].dropna()))
plt.title('log')

fig.add_subplot(223)
plt.hist(np.sqrt(df['agea'].dropna()))
plt.title('Square root')

ax3=fig.add_subplot(224)
plt.hist(1/df['agea'].dropna())
plt.title('Inverse')
plt.show()

#transformations are meh, but still make feature from log

features['log_age'] = np.log(df['agea'])
```

```python
sns.regplot(
    df['agea'],
    y=df['sclmeet'],
    y_jitter=.49,
    order=2,
    scatter_kws={'alpha':0.3},
    line_kws={'color':'black'},
    ci=None
)
plt.show()
features['age_squared']=df['agea'] * df['agea']
```

```python
#reverse code happy variable so higher scores mean less
#happiness and possibly more sadness

features['Sadness'] = max(df['happy']) - df['happy']

sns.regplot(
    df['tvtot'],
    features['Sadness'],
    x_jitter=.49,
    y_jitter=.49,
    scatter_kws={'alpha':0.3},
    line_kws={'color':'black'},
    ci=None
)
plt.xlabel('TV watching')
plt.ylabel('Sadness')
plt.show()
```

```python
#requires sklearn; pip install sklearn
from sklearn import preprocessing
#same as import sklearn.preprocessing

#select numerics only to scale
df_num = df.select_dtypes(include=[np.number]).dropna()

#save column names
names = df_num.columns

#scale then turn resulting numpy array into DF with correct col names
df_scaled = pd.DataFrame(preprocessing.scale(df_num), columns=names)

#new features contain all information of old but on new scale
plt.scatter(df_num['tvtot'], df_scaled['tvtot'])
plt.show()

#look at all matching means and standard deviations
print(df_scaled.describe())
```

```python
#create better indicator for partner. currently coded 1 for living
#with partner and 2 for not living with partner

features['LiveWithPartner'] = np.where(df['partner'] == 1, 1, 0)

#create interaction
features['Sad_Partner'] = features['Sadness'] * features['LiveWithPartner']

#plot interaction
#add tvtot feature to features data frame for plotting
features['tvtot'] = df['tvtot']
sns.lmplot(
    x='Sadness',
    y='tvtot',
    hue='LiveWithPartner',
    data=features,
    scatter=False
)
plt.show()
```
