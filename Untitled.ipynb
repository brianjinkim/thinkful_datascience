```python
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fao_raw = pd.read_csv('../Datasets/FAO.csv', encoding='latin1')
fao_raw.head()
```

```python
fao_raw.info()
```

```python
#drop unneeded rows/cols and rows with missing data
df = fao_raw
df = df.drop(['Area Abbreviation', 'Area Code', 'latitude', 'longitude', 'Unit',
             'Y1961', 'Y1962', 'Y1963', 'Y1964', 'Y1965', 'Y1966', 'Y1967',
              'Y1968', 'Y1969', 'Y1970', 'Y1971', 'Y1972', 'Y1973', 'Y1974',
              'Y1975', 'Y1976', 'Y1977', 'Y1978', 'Y1979', 'Y1980', 'Y1981',
              'Y1982', 'Y1983', 'Y1984', 'Y1985', 'Y1986', 'Y1987', 'Y1988',
              'Y1989', 'Y1990', 'Y1991', 'Y1992', 'Y1993', 'Y1994', 'Y1995',
              'Y1996', 'Y1997', 'Y1998', 'Y1999',], axis=1)
df = df[df['Element'] == 'Food']
df = df.dropna(axis=0)
```

```python
#check for zeros
(df==0).sum(axis=1)
```

```python
#remove rows with 0s
def dropIndex(df, x):
    for val in x:
        df = df.drop(val)
    return df

zero_rows = df[~np.all(df != 0, axis=1)].index
zero_rows = zero_rows.tolist()

df = dropIndex(df, zero_rows)
df.head(10)
```

```python
df.Area.unique()
```

```python
#interested in Republic of Korea, Japan, United States of America, France, Switzerland, United Kingdom
df = df[df['Area'].isin(
    ['Republic of Korea', 'Japan', 'France', 'Switzerland',
     'United States of America','United Kingdom']
)]
df.head()
```

```python
#get rid of element code and element, they're all Food
df.reset_index(drop=True, inplace=True)
df.drop(['Element Code', 'Element'], axis=1, inplace=True)

df.head()
```
